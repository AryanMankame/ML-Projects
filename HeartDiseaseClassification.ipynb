```python
# Spliting our data
X = data.drop('target',axis = 1)
Y = data['target']
```

```python
np.random.seed(42)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
len(X_train) , len(X_test)
```

### Import the models that we will be using since we are doing classification the three classification models used are :
 - Logistic Regression
 - KNearestNeighbour
 - RandomForestClassifier
   


```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    'Logistic Regression' : LogisticRegression(),
    'KNeighbors Regression' : KNeighborsClassifier(),
    'RandomForestClassifier' : RandomForestClassifier()
}
```

```python
def fit_train_score(models,X_train,X_test,Y_train,Y_test):
    np.random.seed(42)
    model_score = {}
    for name,model in models.items():
        model.fit(X_train,Y_train)
        model_score[name] = model.score(X_test,Y_test)
    return model_score
```

```python
model_score = fit_train_score(models,X_train,X_test,Y_train,Y_test)
```

```python
model_score
```

```python
model_df = pd.DataFrame(model_score,index = ["Accuracy"])
```

```python
model_df.plot(kind = 'bar');
```

```python
model_df.T.plot.bar();
```

## Now to improve the accuaracy we will perform hyperparameter tunning
### KNearest Neighbor hyperparameter is n_neighbor

```python
n_range = range(1,21)
train_score = []
test_score = []
for i in n_range:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    train_score.append(knn.score(X_train,Y_train))
    test_score.append(knn.score(X_test,Y_test))
```

```python
plt.plot(n_range,train_score,label = "Train Score")
plt.plot(n_range,test_score,label = "Test Score")
plt.xlabel("No of neighbors")
plt.ylabel("Accuracy")
plt.legend()
plt.xticks(np.arange(1,21,1));
```

```python
print(f"The maximum Accuracy takes place when n_neighbors = {max(test_score)*100:.2f} %")
```

## Hyperparameter tunning using RandomisedSearchCV

## Logistic Regression
Logistic regression does not really have any critical hyperparameters to tune.

Sometimes, you can see useful differences in performance or convergence with different solvers (solver).

solver in [newton-cg, lbfgs, liblinear, sag, saga]
Regularization (penalty) can sometimes be helpful.

penalty in [none, l1, l2, elasticnet]
Note: not all solvers support all regularization terms.

The C parameter controls the penality strength, which can also be effective.

C in [100, 10, 1.0, 0.1, 0.01], 0.1, 0.01]

```python
lg_grid = { 
    "solver" : ['newton-cg','lbfgs','liblinear','sag','saga'],
    "penalty" : ['none','l1','l2','elasticnet'],
    "C" : [10000,1000,100,10,1.0,0.1,0.01,0.001,0.0001]
}
lg_grid1 = {
    "solver" : [ 'liblinear' ],
    "C" : np.logspace(-4,4,20)
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
np.random.seed(42)
lg_tune = RandomizedSearchCV(LogisticRegression(),
                             param_distributions = lg_grid1,
                             cv = 5,
                             n_jobs = -1,
                             n_iter = 20,
                             verbose = True)
lg_tune.fit(X_train,Y_train)
```

```python
lg_tune.best_params_
```

```python
lg_tune.score(X_test,Y_test)
```

```python
rf_grid = {
    "n_estimators" :  np.arange(10,1000,50),
    "max_depth": [None,3,5,10],
    "min_samples_split" : np.arange(2,10,2),
    "min_samples_leaf" : np.arange(1,10,2)
}
```

```python
np.random.seed(42)
rf_tune = RandomizedSearchCV(
    RandomForestClassifier(),
    param_distributions = rf_grid,
    cv = 5,
    verbose = True,
    n_iter = 20,
    n_jobs = -1
)
```

```python
rf_tune.fit(X_train,Y_train)
```

```python
rf_tune.best_params_
```

```python
rf_tune.score(X_test,Y_test)
```

```python
from sklearn.model_selection import GridSearchCV
rf_tune_grid = GridSearchCV(
             LogisticRegression(),
             param_grid=lg_grid1,
             cv = 5,
             verbose = True
            )
```

```python
rf_tune_grid.fit(X_train,Y_train)
```

```python
rf_tune_grid.score(X_test,Y_test)
```

# Evaluating the Model

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report,RocCurveDisplay
```

```python
Y_preds = lg_tune.predict(X_test)
```

```python
acc_score = accuracy_score(Y_test,Y_preds)
print(f"The accuracy score is {acc_score*100:.4f}%")
```

```python
rec_score = recall_score(Y_test,Y_preds)
print(f"The accuracy score is {rec_score*100:.4f}%")
```

```python
f_1_score = f1_score(Y_test,Y_preds)
print(f"The accuracy score is {f_1_score*100:.4f}%")
```

```python
RocCurveDisplay.from_estimator(lg_tune,X_test,Y_test)
```

### Confusion Matrix

```python
conf_mat = confusion_matrix(Y_test,Y_preds)
print(conf_mat)
```

```python
sns.set(font_scale = 1.5)
def plot_conf_matrix(Y_test,Y_preds):
    fig,ax = plt.subplots(figsize = (10,6))
    ax = sns.heatmap(conf_mat,annot=True)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
plot_conf_matrix(Y_test,Y_preds)
```

```python
print(classification_report(Y_test,Y_preds))
```

## Getting all measures based on cross val score

```python
from sklearn.model_selection import cross_val_score
```

```python
# Cross Val Accuracy
cv_acc = cross_val_score(lg_tune,X,Y,cv = 5,scoring = "accuracy")
print(f"The accuracy score is {np.mean(cv_acc)*100:.4f}%")
```

```python
# Cross Val Recall
cv_rec = cross_val_score(lg_tune,X,Y,cv = 5,scoring = "recall")
print(f"The recall score is {np.mean(cv_rec)*100:.4f}%")
```

```python
# Cross Val Precision
cv_prec = cross_val_score(lg_tune,X,Y,cv = 5,scoring = "precision")
print(f"The recall score is {np.mean(cv_prec)*100:.4f}%")
```

```python
cv_metrics = pd.DataFrame({"Accuracy" : np.mean(cv_acc), "Recall" : np.mean(cv_rec), "Precision" : np.mean(cv_prec)},index=[0])
```

```python
cv_metrics.T.plot.bar(legend = False);
```
